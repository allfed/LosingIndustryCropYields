This notebook makes a simple barplot out of the yield reduction values. 

```python
import pandas as pd

# Create the dataframe
data = {'Maize': [-29, -45],
        'Rice': [-33, -37],
        'Soybean': [-17, -38],
        'Wheat': [-35,-46]}
df = pd.DataFrame(data, index=['Phase 1', 'Phase 2']).transpose()

print(df)
```

```python
import matplotlib.pyplot as plt

plt.style.use("https://raw.githubusercontent.com/allfed/ALLFED-matplotlib-style-sheet/main/ALLFED.mplstyle")
```

```python
# Create the plot
fig, ax = plt.subplots()
df.plot(kind='bar', ax=ax)

# Set the title and axis labels
ax.set_ylabel('Projected Yield Reduction [%]')
ax.xaxis.grid(False)
ax.set_xticks(ax.get_xticks(), ax.get_xticklabels(), rotation=0);

plt.savefig("yield_reduction_bar.png", dpi=250)
```
